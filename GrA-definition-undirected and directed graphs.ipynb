# Graph Analysis
### Graph Definition, undirected and directed graphs 
A **graph** is composed of two sets: 
 - Nodes (or vertices) such as cities, people, or webpages 
 - Edges (or links) such as roads, friendships, or hyperlinks

Nodes are the entities. And edges are relationships between entities. 
**Hint:** It is reminded that an edge is defined between two nodes.
There are different kinds oof graphs. One well-known type of graphs is the **undirected graphs**. In an undirected graph, edges have no *direction*. For example, freindships on Facebook have no direction.
<br> In contrast, we have **directed graphs** in which edges have direction. For instance, the following in Twitter is a directed relationship. IF A follows B, it doesn't mean that B follows A. B may follow or may not follow A.

```python
# Import required modules
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create an undirected graph
G = nx.Graph()

# Add nodes
G.add_node("Alice")
G.add_node("Bob")
G.add_node("Charlie")

# Or add multiple nodes at once
G.add_nodes_from(["Diana", "Eve"])

# Add edges (friendships)
G.add_edge("Alice", "Eve")
G.add_edge("Bob", "Charlie")
G.add_edge("Bob","Diana")
G.add_edge("Bob","Eve")
G.add_edge("Alice", "Charlie")
G.add_edge("Diana", "Eve")

# Draw the graph
nx.draw(G, with_labels=True, node_color='lightblue', 
        edge_color='gray', node_size=1600, font_size=12)
plt.title("Undirected Graph: Facebook-style Friendships")
plt.show()
```

```python
# Create a directed graph
D = nx.DiGraph()

# Add nodes
D.add_nodes_from(["Alice", "Bob", "Charlie"])

# Add directed edges (follows)
D.add_edge("Alice", "Bob")     # Alice follows Bob
D.add_edge("Bob", "Charlie")   # Bob follows Charlie
D.add_edge("Charlie", "Alice") # Charlie follows Alice

# Draw the graph
nx.draw(D, with_labels=True, node_color='lightcoral', edge_color='gray', 
        node_size=1500, font_size=12, arrows=True, arrowstyle='->', arrowsize=20)
plt.title("Directed Graph: Twitter-style Following")
plt.show()
```
